# City Analysis

## Objectives

- Identify the city with the highest number of restaurants in the dataset.

- Calculate the average rating for restaurants in each city.

- Determine the city with the highest average rating.


## Methodology
- Top 10 Cities with Most Restaurants
- Top 10 cities with the highest average ratings
- Highlight City with the Highest Average Rating

```python
# Loading Libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Cleaned Dataset

df_cleaned = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Data_Analyst\Cognifyz_Test\Dataset\Cleaned_Dataset.csv')
```

```python
# Count the number of restaurants in each city
city_counts = df_cleaned['City'].value_counts()

# Get the city with the highest number of restaurants
city_highest_restaurants = city_counts.idxmax()
highest_restaurant_count = city_counts.max()

print(f"City with the highest number of restaurants: {city_highest_restaurants} ({highest_restaurant_count} restaurants)")
```

```python
# Calculate average rating for each city

average_rating_per_city = df_cleaned.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)

print("Average ratings for each city:")
print(average_rating_per_city.head())
```

```python
# Get the city with the highest average rating

city_highest_rating = average_rating_per_city.idxmax()
highest_average_rating = average_rating_per_city.max()

print(f"City with the highest average rating: {city_highest_rating} ({highest_average_rating:.2f})")
```

```python
# Top Cities Bar Chart

# Top 10 cities with the most restaurants
top_cities = city_counts.head(10)

plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar', color='skyblue')
plt.title("Top 10 Cities with Most Restaurants", fontsize=16)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Restaurants", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Top 10 cities with the highest average ratings
top_avg_ratings = average_rating_per_city.head(10)
plt.figure(figsize=(10, 6))
top_avg_ratings.plot(kind='bar', color='lightgreen')
plt.title("Top 10 Cities with Highest Average Ratings", fontsize=16)
plt.xlabel("City", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
